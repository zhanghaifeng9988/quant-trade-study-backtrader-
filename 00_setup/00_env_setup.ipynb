## 2. 创建虚拟环境（推荐）

虚拟环境可以让不同项目的依赖互不干扰。

### 方式一：conda（推荐）
```bash

conda activate quant-tutorial
```

### 方式二：pip + venv
```bash
python -m venv venv
# Windows
venv\Scripts\activate
# macOS/Linux
source venv/bin/activate

pip install -r requirements.txt
```

## 3. 验证安装

````python
# 运行此单元格验证核心包是否安装成功
packages = [
    ('numpy', 'np'),
    ('pandas', 'pd'),
    ('matplotlib', 'plt'),
    ('yfinance', 'yf'),
    ('scipy', 'sp'),
]

all_ok = True
for pkg, alias in packages:
    try:
        mod = __import__(pkg)
        version = getattr(mod, '__version__', 'unknown')
        print(f'✅ {pkg:<15} v{version}')
    except ImportError:
        print(f'❌ {pkg:<15} 未找到，请运行: pip install {pkg}')
        all_ok = False

print('\n🎉 环境配置完成！' if all_ok else '\n⚠️  请安装缺少的包后重启 kernel')
````

## 4. Jupyter 基本操作

| 快捷键 | 功能 |
|--------|------|
| `Shift + Enter` | 运行当前单元格，移到下一个 |
| `Ctrl + Enter` | 运行当前单元格，不移动 |
| `A` (命令模式) | 在上方插入新单元格 |
| `B` (命令模式) | 在下方插入新单元格 |
| `D D` (命令模式) | 删除当前单元格 |
| `M` (命令模式) | 切换为 Markdown 单元格 |
| `Y` (命令模式) | 切换为代码单元格 |
| `Esc` | 进入命令模式 |

> 💡 **提示**：单元格边框蓝色 = 命令模式，绿色 = 编辑模式

````python
# 练习：在 Jupyter 中绘制第一张图
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, color='steelblue', linewidth=2)
plt.title('你的第一张图表 🎉', fontsize=14)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print('如果图表正常显示，说明 matplotlib 工作正常！')
````
